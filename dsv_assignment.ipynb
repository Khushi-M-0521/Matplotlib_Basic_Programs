```python
# 1a. Write a Python program to demonstrate how to draw a Stacked Bar Plot using Matplotlib

import matplotlib.pyplot as plt

# Data for the plot
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = range(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

# Plot the data
plt.bar(ind, menMeans, width, label='Men')
plt.bar(ind, womenMeans, width, bottom=menMeans, label='Women')

# Add title and labels
plt.title('Scores by group and gender')
plt.xlabel('Group')
plt.ylabel('Score')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#1b. Write a Python program to demonstrate how to draw a Scatter Plot, Bubble Plot using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

# Generate 50 random points
x = np.random.rand(50)
y = np.random.rand(50)
area = np.random.rand(50) * 50  # size of the bubbles

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Bubble plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=area, c='blue', alpha=0.5)
plt.title('Bubble Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
# 2a. Write a Python program to demonstrate how to draw a Histogram Plot, Box Plot using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Generate a random dataset
data = np.random.randn(500)

# Histogram plot
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, edgecolor='black')
plt.title('Histogram Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=True)
plt.title('Box Plot')
plt.xlabel('Value')
plt.show()
```

```python
#2b. Write a Python program to demonstrate how to draw a Pie Chart using Matplotlib. 
import matplotlib.pyplot as plt

# Data for the pie chart
labels = 'A', 'B', 'C', 'D', 'E', 'F', 'G'
sizes = [10, 15, 20, 12, 18, 25, 10]
colors = ['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink']
explode = (0.1, 0, 0, 0, 0, 0, 0)  # explode the first slice

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#3a. Write a Python program to illustrate Linear Plotting using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = 2*x + 3

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', linewidth=2, label='y = 2x + 3')
plt.title('Linear Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#3b. Write a Python program to illustrate liner plotting with line formatting using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y1 = 2*x + 3
y2 = 3*x - 2
y3 = x**2 - 4*x + 1

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='blue', linewidth=2, label='y = 2x + 3', linestyle='solid')
plt.plot(x, y2, color='red', linewidth=2, label='y = 3x - 2', linestyle='dashed')
plt.plot(x, y3, color='green', linewidth=2, label='y = x^2 - 4x + 1', linestyle='dotted')
plt.title('Linear Plot with Line Formatting')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#4. Write a Python program which explains uses of customizing seaborn plots with Aesthetic functions.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=tips)

# Add a title and labels
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")

# Customize the plot with aesthetic functions
sns.set_style("whitegrid")  # change the plot style
sns.despine()  # remove the top and right spines
plt.tick_params(direction="in")  # change the direction of the tick marks

# Show the plot
plt.show()
```

```python
#5a. Write a Python program to explain working with using Annotations and bokeh line graph Legends.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, Label
from bokeh.embed import components
from bokeh.models.annotations import Legend

# Create some data
x = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a plot
p = figure(title="Sine and Cosine", x_axis_label='x', y_axis_label='y')

# Add lines to the plot
line1 = p.line(x, y1, color='red', legend_label='Sine')
line2 = p.line(x, y2, color='blue', legend_label='Cosine')

# Add annotations
p.circle([2 * np.pi], [0], size=10, color='red', legend_label='Max Sine')
p.circle([np.pi], [0], size=10, color='blue', legend_label='Max Cosine')

# Add a legend
legend = Legend(items=[("Sine", [line1]), ("Cosine", [line2]), ("Max Sine", [p.circle([2 * np.pi], [0], size=10, color='red')]), ("Max Cosine", [p.circle([np.pi], [0], size=10, color='blue')])], location="center")
p.add_layout(legend, 'right')

# Add a hover tool
hover = HoverTool(tooltips=[("x", "$x"), ("y", "$y")], mode="vline")
p.add_tools(hover)

# Show the plot
show(p)
```

```python
#5b. Write a Python program to draw radar chart using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Speed', 'Accuracy', 'Memory', 'Graphics', 'CPU']
data = [90, 80, 70, 80, 90]

# Create a polar plot
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(polar=True)

# Set the radar chart categories
ax.set_thetagrids(np.linspace(0, 360, len(categories), endpoint=False), categories)

# Set the radar chart values
ax.plot(data)

# Fill the area
ax.fill(data, 'r', alpha=0.2)

# Set the radar chart title
ax.set_title('Radar Chart Example', va='bottom')

# Show the plot
plt.show()
```

```python
#6. Write a Python program to draw 3D Plots using Plotly Libraries.
import plotly.graph_objects as go
import numpy as np

# Create a 3D grid of points
x = np.linspace(-10, 10, 20)
y = np.linspace(-10, 10, 20)
x, y = np.meshgrid(x, y)

# Create a 3D surface
z = x**2 + y**2

# Create a 3D plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Viridis')])

# Add a title and axis labels
fig.update_layout(title='3D Surface Plot', scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))

# Show the plot
fig.show()
```

```python
#7a. Write a Python program to draw Time Series using Plotly Libraries.
import plotly.graph_objects as go
import pandas as pd

# Create a sample time series data
data = {
    'Date': ['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04', '2020-01-05'],
    'Value': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a time series plot
fig = go.Figure(data=[go.Scatter(x=df['Date'], y=df['Value'])])

# Add a title and axis labels
fig.update_layout(title='Time Series Plot', xaxis_title='Date', yaxis_title='Value')

# Show the plot
fig.show()
```

```python
#7b.  Write a Python program for creating Maps using Plotly Libraries.
import plotly.graph_objects as go

# Create a sample data
countries = ['USA', 'CAN', 'MEX', 'BRA', 'IND']
values = [100, 50, 20, 80, 120]

# Create a map plot
fig = go.Figure(data=go.Choropleth(
    locations=countries,
    z=values,
    text=countries,
    colorscale='Blues',
    colorbar_title='Values'
))

# Add a title and axis labels
fig.update_layout(title='Map Plot', geo_scope='world')

# Show the plot
fig.show()


```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a random number dataset
np.random.seed(0)
data = np.random.rand(10, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)

# Draw the correlogram
sns.correlogram(corr, ax=ax[1])

plt.show()


```
